# LINEAR ALGEBRA

Suppose there a sheriff is travelling at 180 Kmph and robber is travelling at 150 Kmph with a 5 minute headstart, How long does it take for the Sheriff to catch the robber?

```python
#Since the distance travelled by the sheriff and robber are the same,
#We can set up equations as:
import numpy as np
import matplotlib.pyplot as plt
time = np.linspace(0, 40, 100)
distance_sheriff = (180/60) * (time - 5)
distance_robber = (150/60) * time

#Plot the graph to find the point at which the two lines intersect. This point is where the sheriff catches the robber
fig, ax = plt.subplots()
plt.title('Distance to catch the robber')
plt.xlabel('Time')
plt.ylabel('Distance')
ax.set_xlim([0, 40])
ax.set_ylim([0, 100])
ax.plot(time, distance_sheriff, c = 'red')
ax.plot(time, distance_robber, c = 'green')
```
